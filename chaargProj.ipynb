```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# read in chaarg data
data = pd.read_csv("./chaargData.csv")
data.head()
```

```python
# plot data to find trends

# which states have the most current chaarg chapters
stateCounts = data['State'].value_counts()
stateCounts.plot(kind = 'bar')
plt.title('CHAARG Chapters Per State')
plt.xlabel('States')
plt.ylabel('Count')
plt.show()
```

```python
# how many total states is chaarg in?
numCHAARGStates = len(stateCounts)
print(f"CHAARG chapters are in {numCHAARGStates} states!")
```

```python
# which of the three semesters were memberships the highest?

cleanData = data.dropna()
cleanData = cleanData[['Chapter Name', 'Fall 2022', 'Spring 2023', 'Fall 2023']]
plt.bar(['Fall 2022', 'Spring 2023', 'Fall 2023'], [cleanData['Fall 2022'].sum(), cleanData['Spring 2023'].sum(), cleanData['Fall 2023'].sum()])
plt.title("Total CHAARG Members Per Semester")
plt.xlabel("Semester")
plt.ylabel("Number Members")
```

```python
# let's show this with a different type of plot as well
plt.boxplot([cleanData['Fall 2022'], cleanData['Spring 2023'], cleanData['Fall 2022']], labels=['Fall 2022', 'Spring 2023', 'Fall 2023'])
plt.title('Average CHAARG Memberships Per Semester')
plt.xlabel('Semester')
plt.ylabel('Memberships')
```

```python
fall2022Mean = np.mean(cleanData['Fall 2022'])
print(f'The mean number of members for all chapters in Fall 2022 is {round(fall2022Mean, 2)}')
spring2023Mean = np.mean(cleanData['Spring 2023'])
print(f'The mean number of members for all chapters in Spring 2023 is {round(spring2023Mean, 2)}')
fall2023Mean = np.mean(cleanData['Fall 2023'])
print(f'The mean number of members for all chapters in Fall 2023 is {round(fall2023Mean, 2)}')
```
